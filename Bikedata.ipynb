```python
import pandas as pd
import datetime as dt 
```

```python
bikedata=pd.read_csv("C:/users/lalit/learnpanda/201902-citibike-tripdata.csv",low_memory=False)
```

```python
bikedata.head()
```

```python
bikedata['starttime']= bikedata.starttime.apply(lambda d:dt.datetime.strptime(d, '%Y-%m-%d %H:%M:%S.%f'))
```

```python
bikedata['startdate']=bikedata.starttime.apply(lambda d: d.date())
bikedata['start_time']=bikedata.starttime.apply(lambda d: d.time())
```

```python
bikedata['stoptime']= bikedata.stoptime.apply(lambda d:dt.datetime.strptime(d, '%Y-%m-%d %H:%M:%S.%f'))
```

```python
bikedata['stopdate']=bikedata.stoptime.apply(lambda d : d.date())
bikedata['stop_time']=bikedata.stoptime.apply(lambda d : d.time())
```

```python
bikedata.head()
```

```python
def gender_dt(D):
    if (D == 1):
        return 'Male'
    elif (D== 2) :
        return 'Female'
    else :
        return 'Unknown'
```

```python
bikedata['gender']=bikedata.gender.apply(lambda d : gender_dt(d))
```

```python
bikedata['age']=bikedata['birth year'].apply(lambda d: (dt.datetime.now().year- d) if(dt.datetime.now().year-d < 80) else 0)
```

```python
bikedata.head()
```

```python
bikedata['start station id'].fillna(0, inplace=True)
bikedata['end station id'].fillna(0, inplace=True)
bikedata['start station id']=bikedata['start station id'].astype(int)
bikedata['end station id']= bikedata['end station id'].astype(int)
bikedata['start station name'].fillna('UNKNOWN', inplace=True)
bikedata['end station name'].fillna('UNKNOWN', inplace=True)
```

```python
bikedata.head()
```

```python
bikedata.to_csv("C:\DataAnalytics\Bikedata0219.csv")
```

```python
path_df1=bikedata[['start station id','start station name','start station latitude','start station longitude']].copy()
```

```python
path_df1['path']=bikedata['start station id'].astype(str)+ '__'+ bikedata['end station id'].astype(str)
```

```python
path_df1.head()
```

```python
path_df1.columns = ['Stationid','Station_name', 'latitude',
      'longitude','path']
path_df1['type']='Start'
```

```python
path_df1
```

```python
path_df1['cnt']=path_df1.index
```

```python
path_df1
```

```python
pathdf1= path_df1.rename(columns:{'start station id':'Station id','start station name': 'station name','start station latitude':'latitude','start station longitude':'longitude','path':'path'})
```

```python
path_df2=bikedata[['end station id','end station name','end station latitude','end station longitude']].copy()
```

```python
path_df2['path']=bikedata['start station id'].astype(str)+ '__'+ bikedata['end station id'].astype(str)
```

```python
path_df2.columns = ['Stationid','Station_name', 'latitude',
      'longitude','path']
path_df2['type']='Destination'
path_df2['cnt']=path_df1.index
```

```python
path_final=path_df1.append(path_df2,ignore_index=True)
```

```python
path_final =path_final.groupby(['path','Stationid','Station_name','latitude','longitude','type']).agg('count')
```

```python
path_final.reset_index(inplace=True)
```

```python
path_final=path_final[path_final.path != '0__0']
```
